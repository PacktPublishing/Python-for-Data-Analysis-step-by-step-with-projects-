# Pivoting tables

```python
import pandas as pd
student = pd.read_csv('student.csv')
```

```python
student.info()
```

```python
student
```

```python
pd.pivot_table(student, 
               index='sex')
```

```python
pd.pivot_table(student, 
               index=['sex', 'internet']
              )
```

```python
pd.pivot_table(student, 
               index=['sex', 'internet'], 
               values='score')
```

```python
pd.pivot_table(student, 
               index=['sex', 'internet'], 
               values='score', 
               aggfunc='mean')
```

```python
pd.pivot_table(student, 
               index=['sex', 'internet'], 
               values='score', 
               aggfunc='median')
```

```python
pd.pivot_table(student, 
               index=['sex', 'internet'], 
               values='score', 
               aggfunc=['min', 'mean', 'max'])
```

```python
pd.pivot_table(student, 
               index=['sex', 'internet'], 
               values='score', 
               aggfunc='mean',
               columns='studytime'
              )
```

```python
student[(student['sex']=='M') & (student['internet']=='no') & (student['studytime']=='4. >10 hours')]
```

```python
pd.pivot_table(student, 
               index=['sex', 'internet'], 
               values='score', 
               aggfunc='mean',
               columns='studytime',  
               fill_value=-999)
```

```python
pd.pivot_table(student, 
               index=['sex', 'internet'], 
               values=['score', 'age'],
               aggfunc='mean', 
               columns='studytime', 
               fill_value=-999)
```

```python
pd.pivot_table(student, 
               index=['sex'], 
               values='score', 
               aggfunc='mean',                
               columns=['internet', 'studytime'],
               fill_value=-999)
```

```python
pd.pivot_table(student, 
               index='familysize',
               values='score', 
               aggfunc='mean',
               columns='sex'
              )
```

```python
pd.pivot_table(student, 
               index='familysize',
               values='score', 
               aggfunc='mean',
               columns='sex', 
               margins=True,
               margins_name='Average score total')
```

```python
student[student['sex']=='F'].mean()
```

```python
pd.pivot_table(student, 
               index='studytime', 
               values=['age', 'score'], 
               aggfunc={'age': ['min', 'max'],
                        'score': 'median'}, 
               columns='sex')
```

```python
pd.pivot_table(student,  
               index='studytime', 
               values='score',
               aggfunc=lambda s: s.max() - s.min(),
               columns='sex'
               )
```
